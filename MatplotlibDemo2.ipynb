```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rand_numbers=np.arange(10)
print(rand_numbers)
```

```python
rand_numbers_square=rand_numbers**2
rand_numbers_cube=rand_numbers**3
```

```python
#How to plot using matplotlib
plt.plot(rand_numbers,color="green")
plt.show()
```

```python
plt.plot(rand_numbers,rand_numbers_square,color="orange",label="square of x")
plt.plot(rand_numbers,rand_numbers_cube,color="green",label="cube of x")
plt.legend()
plt.xlabel("rand_numbers")
plt.ylabel("values")
plt.show()
```

```python
#plot a scatter plot
plt.scatter(rand_numbers,rand_numbers_square,marker="^",label="square of random_numbers")
plt.legend()
plt.xlabel("x")
plt.ylabel("x square")
plt.show()
```

```python
#Plotting Bar Charts
values=np.random.randint(0,10,8)
indices=np.arange(8)

print(values)
print(indices)
```

```python
plt.bar(indices,values)
plt.show()
```

```python
#changing the color
plt.bar(indices,values,color="green")
plt.show()
```

```python
#Adding more bars int he same plot
values2=np.random.randint(0,10,8)
indices2=np.arange(8)

plt.bar(indices,values,color="red",label="red_label",width=0.3)
plt.bar(indices2+0.3,values2,color="blue",label="blue_label",width=0.3)
plt.legend()
plt.show()
```

```python
#How to plot a histogram
#Lets first make a data which belongs to normal distribution.Normal Distribution is made for continuous random variables

#Explain random variable,Discrete random variable and continous random variable
# The first parameter means mean ,second parameter means std,thord parameter means total number of values and all the values
# will belong to the normal distribution
data=np.random.normal(1,10,30)
print(data)
```

```python
plt.hist(data,bins=5)
```

```python
#Lets set the density parameter to true
plt.hist(data,bins=5,density="True")
```

```python
np.histogram(data,bins=5,density="True")
```
