```python
def distance(p1,p2):
    return np.sqrt(np.sum((p1-p2)**2))
def knn(xtrain,ytrain,point,k=5):
    all_dist=[]
    for i in range(xtrain.shape[0]):
        dist=distance(xtrain[i],point)
        all_dist.append((dist,ytrain[i]))
    all_dist=sorted(all_dist)
    all_dist=np.array(all_dist)
    all_dist=all_dist[:k]  
    print(all_dist)
    freq=np.unique(all_dist[:,1],return_counts=True)
    print(freq)
    idx=np.argmax(freq[1])
    pred=freq[0][idx]
    return pred
```
