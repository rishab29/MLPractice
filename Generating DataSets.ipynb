# What is a norm of a Vector?
-In linear algebra, functional analysis, and related areas of mathematics, a norm is a function that assigns a strictly positive length or size to each vector in a vector spaceâ€” except for the zero vector, which is assigned a length of zero. 
There are various norms like l1,l2,l3, and so on.

The norms are used in regularization.

# Calculate the norm of a vector

```python
#l2 norm-sum of squares of all values and then its underoot.
x=[1,-3,4]
l2_norm=np.linalg.norm(x)
print(l2_norm)
```

```python
#l1 norm- sum of absolute values of all elements in vector. 
l1_norm=np.linalg.norm(x,ord=1)
print(l1_norm)
```

```python
# infinity norm-It is the highest absolute value in vector
inf_norm=np.linalg.norm(x,ord=np.inf)
print(inf_norm)
```

# Data for classification

```python
#Binary Classification
means=np.array([2,4])
covariance=np.array([[2,-.8],[-.8,1]])
x1=np.random.multivariate_normal(means,covariance,300)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
means2=np.array([-1,-3])
covariance2=np.array([[1,0],[0,2]])
x2=np.random.multivariate_normal(means2,covariance2,400)
plt.scatter(x2[:,0],x2[:,1])
plt.show()
```

```python
X=np.vstack((x1,x2))
Y=np.zeros(700)
Y[:300]=1
```

```python
#one way to genrate label them with different colors
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
#another way to do the same is
print(X.shape)
```

```python
for i in range(700):
    if Y[i]==0:
        plt.scatter(X[i][0],X[i][1],color="red")
    else:
        plt.scatter(X[i][0],X[i][1],color="green")
        
plt.show()        
```
