```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
x=np.array([1,2,3,4])
```

```python
plt.plot(x,"--",label="x values")
plt.xlabel("Indices")
plt.ylabel("Values")
plt.title("Value of X")
plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3,4],[10,20,30,40],marker="x")
plt.scatter([10,20,3],[40,50,60])
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot([1,2,3])
plt.xlabel("xvalues")
plt.ylabel("yvalues")
plt.title("Plot 1")

plt.subplot(2,1,2)
plt.plot([4,5,6],"--",c="red")
plt.xlabel("xvalues")
plt.ylabel("yvalues")
plt.title("Plot 2")

plt.subplots_adjust(right=2.2,top=1.8)
plt.show()
```
